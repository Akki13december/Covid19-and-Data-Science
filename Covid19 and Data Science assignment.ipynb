```python
dataset2.Negative = dataset2.Negative.fillna('0') 
dataset2.Positive = dataset2.Positive.fillna('0') 
```

```python
dataset1.rename(columns = {'State/UnionTerritory':'State'},inplace=True)
```

```python
dataset1
```

```python
dataset1['State'].unique()
```

```python
dataset1=dataset1.replace('Telengana','Telangana')
dataset1=dataset1.replace('Telengana***','Telangana')
dataset1=dataset1.replace('Telangana***','Telangana')
dataset1=dataset1.replace('Maharashtra***','Maharashtra')
dataset1=dataset1.replace('Chandigarh***','Chandigarh')
dataset1=dataset1.replace('Punjab***','Punjab')
```

```python
dropn_indexnames =dataset1[(dataset1['State'] == 'Cases being reassigned to states')].index 
dataset1.drop(dropn_indexnames,inplace=True)
```

```python
df = dataset1.head(5)
df1 = df.sort_values(by='Confirmed', ascending=False).head(5)
```

```python
dataset2 = dataset2.astype({'Positive': 'int32'})
```

```python
dataset2.dtypes
```

```python
x = ['Deaths', 'Cured', 'Confirmed', 'Active']
h = [289, 1490, 28, 1462]
c = ['black', 'yellow', 'red', 'blue']
plt.bar(x, height = h, color = c)
plt.show()
```

```python
a4_dims = (13, 7)
fig, ax = pyplot.subplots(figsize=a4_dims)
barp = sns.barplot(data=df1,x='State',y='Confirmed',ax=ax)
barp.set_yticklabels(labels=(barp.get_yticks()*1).astype(int))
plt.show()
```

```python
df1['Fatality-Ratio'] = df1['Deaths']/df1['Confirmed']
a4_dims = (15, 7)
fig, ax = pyplot.subplots(figsize=a4_dims)
plt.title('Death-Ratio of Top 5 Contaminated States',size=15)
sns.pointplot(data=df1,x='State',y='Fatality-Ratio',ax=ax,color='Black')
plt.show()
```

```python
df1['Cure-Ratio'] = df1['Cured']/df1['Confirmed']
a4_dims = (15, 7)
fig, ax = pyplot.subplots(figsize=a4_dims)
plt.title('Cure-Ratio of Top 5 Contaminated States',size=15)
sns.pointplot(data=df1,x='State',y='Cure-Ratio',ax=ax,color='Yellow')
plt.show()
```

```python
Delhi = dataset1[dataset1.State == 'Delhi']
Delhi
```

```python
req = Delhi[Delhi['Date'].isin(['09/03/20','09/04/20','09/05/20','09/06/20','09/07/20','09/08/20','09/09/20','09/10/20','09/11/20','09/12/20'])]
```

```python
req
```

```python
a4_dims = (13, 7)
fig, ax = pyplot.subplots(figsize=a4_dims)
pp = sns.lineplot(data=req,x='Date',y='Confirmed',ax=ax,color='orange')
pp = sns.lineplot(data=req,x='Date',y='Cured',ax=ax,color='green')
pp = sns.lineplot(data=req,x='Date',y='Deaths',ax=ax,color='red')
pp.set_yticklabels(labels=(pp.get_yticks()*1).astype(int))
fig.legend(labels=['Confirmed','Cured','Deaths'])
plt.title('Vulnerability summary of Delhi',size=15)
plt.show()
```

```python
test_delhi = dataset2[dataset2.State == 'Delhi']
reqe = test_delhi[test_delhi['Date'].isin(['2020-04-05','2020-05-05','2020-06-05','2020-07-05','2020-08-05','2020-09-05','2020-10-05','2020-11-05','2020-12-05'])]
a4_dims = (13, 7)
fig, ax = pyplot.subplots(figsize=a4_dims)
lp = sns.lineplot(x='Date',y='TotalSamples',data=reqe,ax=ax,color='green')
lp.set_yticklabels(labels=(lp.get_yticks()*1).astype(int))
plt.title('Testing in Delhi',size=15)
plt.show()
```

```python
a4_dims = (13, 7)
fig, ax = pyplot.subplots(figsize=a4_dims)
reqe['positive/samples'] = (reqe['Positive']/reqe['TotalSamples'])
df =reqe.iloc[1:7]
sns.lineplot(x='Date',y='positive/samples',ax=ax,data=df)
plt.title('Positive Cases per Samples in Delhi',size=15)
plt.show()
```
