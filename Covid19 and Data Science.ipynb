```python
df = dataset1.head(5)
df1 = df.sort_values(by='Confirmed', ascending=False).head(5)
```

```python
x = ['Deaths', 'Cured', 'Confirmed', 'Active']
h = [289, 1490, 28, 1462]
c = ['black', 'yellow', 'red', 'blue']
plt.bar(x, height = h, color = c)
plt.show()
```

```python
a4_dims = (13, 7)
fig, ax = pyplot.subplots(figsize=a4_dims)
barp = sns.barplot(data=df1,x='State',y='Confirmed',ax=ax)
barp.set_yticklabels(labels=(barp.get_yticks()*1).astype(int))
plt.show()
```

```python
df1['Fatality-Ratio'] = df1['Deaths']/df1['Confirmed']
a4_dims = (15, 7)
fig, ax = pyplot.subplots(figsize=a4_dims)
plt.title('Death-Ratio of Top 5 Contaminated States',size=15)
sns.pointplot(data=df1,x='State',y='Fatality-Ratio',ax=ax,color='Black')
plt.show()
```

```python
df1['Cure-Ratio'] = df1['Cured']/df1['Confirmed']
a4_dims = (15, 7)
fig, ax = pyplot.subplots(figsize=a4_dims)
plt.title('Cure-Ratio of Top 5 Contaminated States',size=15)
sns.pointplot(data=df1,x='State',y='Cure-Ratio',ax=ax,color='Yellow')
plt.show()
```
